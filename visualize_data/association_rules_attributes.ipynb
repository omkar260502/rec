## Association rules (2 attributes)

```python
# initialize dict of attributes co-occurences
dic = dict()
for i in range(len(mlb.classes_)):
    for j in range(i+1, len(mlb.classes_)):
        att1 = mlb.classes_[i]
        att2 = mlb.classes_[j]
        dic[(att1,att2)] = 0
```

```python
columns = list(map(str,one_hot_dataframe.columns.values))
```

```python
def assoc_rule(row):
    for i in range(len(columns)):
        col1 = columns[i]
        if row[col1] == 1:
            for j in range(i+1, len(columns)):
                col2 = columns[j]
                if row[col2] == 1:
                    dic[(col1,col2)] += 1
```

```python
# get total count of single attributes
attributes_count = one_hot_dataframe.sum()
```

```python
attributes_count
```

```python
confidences = []
for i in range(len(columns)):
        a1 = columns[i]
        for j in range(i+1, len(columns)):
            a2 = columns[j]
            c1 = attributes_count[a1]
            c2 = attributes_count[a2]
            cd = dic[(a1,a2)]
            conf1 = cd / c1
            conf2 = cd / c2
            confidences.append( (a1,a2, c1, c2, cd, conf1, conf2, conf1 + conf2 ) )
            #confidences.append( (a2,a1, c2, c1, cd, cd / c2) )
```

```python
confidences[:20]
```

```python
assoc_rule_df = pd.DataFrame(confidences, columns=['att1','att2','count_att1','count_att2','count_both','confidence12','confidence21','confidence_sum'], dtype=float)
assoc_rule_df = assoc_rule_df.sort_values('confidence_sum', ascending=False)
```
