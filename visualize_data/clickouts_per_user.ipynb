# How many clickout has every user?

```python
import sys
import os

os.chdir("../")

import pandas as pd
import matplotlib
```

```python
df_train = pd.read_csv('dataset/original/train.csv')
df_train = df_train[df_train['action_type'] == 'clickout item']
df_train = df_train[~df_train['reference'].isnull()]
```

```python
df_test = pd.read_csv('dataset/original/test.csv')
df_test = df_test[df_test['action_type'] == 'clickout item']
df_test = df_test[~df_test['reference'].isnull()]
```

```python
df_train = pd.concat([df_train, df_test], ignore_index=True)
```

```python
df_train
```

```python
user_clickouts = df_train.groupby('user_id').size().reset_index(name='size')
```

```python
user_clickouts
```

```python
len(user_clickouts[user_clickouts['size'] > 1])/len(user_clickouts)
```

## Turns out that 44% of the user have at least two clickouts.
