```python
def find(df):
    """
    Find Indices of last row of session
    """
    temp_df = df.copy()
    indices = []
    cur_ses = ''
    cur_user = ''
    for idx in tqdm(temp_df.index.values[::-1]):
        ruid = temp_df.at[idx, 'user_id']
        rsid = temp_df.at[idx, 'session_id']
        if (ruid != cur_user or rsid != cur_ses):
            # append the original index
            indices.append(idx)
            cur_user = ruid
            cur_ses = rsid
    return indices[::-1]
```
