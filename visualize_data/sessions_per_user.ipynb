```python
import sys
import os
os.chdir("../")

import pandas as pd
import matplotlib
import data
```

```python
train_df = data.full_df()
```

```python
df = train_df#.head(200000)
```

```python
df
```

```python
def func(x):
    if len(x) > 1:
        return len(x)
    else:
        return pd.DataFrame()
```

```python
df = df[['user_id', 'session_id']].drop_duplicates()
```

```python
grouped = df.groupby(['user_id']).apply(func)
grouped
```

```python
user_more_than_one_session = len(set(list(grouped['user_id'].values)))
user_more_than_one_session
```

```python
all_users = len(set(list(df['user_id'].values)))
all_users
```

```python
user_more_than_one_session/all_users
```

```python
def func1(x):
    return len(x)    
```

```python
grouped1 = df.groupby(['user_id']).apply(func1)
grouped1
```

```python
sum(grouped1.values)/len(grouped1.values)
```
